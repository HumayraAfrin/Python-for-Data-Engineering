## <b>Get unique values from list using map and set</b>

The map() function is used to apply a given function to every item of an iterable, such as a list or tuple, and returns a map object (which is an iterator).

```python
list(map(lambda order: order.split(',')[3], orders_1))
```

```python
len(list(map(lambda order: order.split(',')[3], orders_1)))
```

A set in Python is a built-in data type that stores an unordered collection of unique elements. Sets are mutable, meaning items can be added or removed after the set is created, but the elements themselves must be immutable. Sets are commonly used for membership testing, removing duplicates, and performing mathematical set operations.

```python
set(map(lambda order: order.split(',')[3], orders_1))
```

Python offers two primary methods for sorting: sorted() and list.sort().

## sorted() Function:
- Returns a new sorted list from any iterable (e.g., lists, tuples, strings).
- Does not modify the original iterable.
- Accepts optional key and reverse arguments for custom sorting.

When it comes to list comprehension, we need to make sure we actually specify the logic in square brackets.
Then we can say for col in column details because column details is of type list.

```python
# Display column names using List Comprehension
[col['column_name'] for col in column_details]
```

```python
# Display column names using Map
list(map(lambda col: col['column_name'], column_details))
```

```python
sorting_key = input('Enter sorting key: ') #data_type
columns = sorted(column_details, key=lambda col: col[sorting_key])
columns
```

```python
[col['data_type'] for col in columns]
```

```python
def get_column_names(schemas, ds_name, sorting_key='data_type'):
    column_details = schemas[ds_name]
    columns = sorted(column_details, key=lambda col: col[sorting_key])
    return [col['data_type'] for col in columns]
```

```python
get_column_names(schemas, 'orders')
```
