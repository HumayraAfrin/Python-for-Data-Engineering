## Aggregations using Pandas Dataframe

Here are the tasks related to aggregations.

- Get count by order status.
- Get count by order month and then by order status. We need to generate order month using order date using apply on dataframe.

```python
orders.groupby?
```

```python
orders.columns
```

```python
## Group by 'order_status' and calculate the count of 'order_id' for each group
orders.groupby('order_status')['order_id'].agg(Counts = 'count')
```

```python
#Get count by order month
orders.groupby('order_date').agg('count')
```

### <b>apply() method:</b> This function acts as a map() function in Python. It takes a function as an input and applies this function to an entire DataFrame. If you are working with tabular data, you must specify an axis you want your function to act on ( 0 for columns; and 1 for rows).

```python
#Adding new column order_month using apply() and lambda functions
#axis = 1 indicates that it will look for column by name
orders['order_month'] = orders.apply(lambda order: order.order_date[:7], axis = 1)
orders
```

```python
#Get count by order month and then by order status.
orders.\
    groupby(['order_month', 'order_status'])['order_id'].\
        agg(Counts = 'count')
```

The reset_index() function in Python, part of the Pandas library, is used to reset the index of a DataFrame or Series to the default integer index (0, 1, 2, ...). This function is particularly useful after operations like filtering, sorting, or grouping that may result in a non-sequential or custom index.

```python
orders.\
    groupby(['order_month', 'order_status'])['order_id'].\
        agg(Counts = 'count').\
            reset_index()  # <-- used for system generated index
```

## Joining Data in Pandas Dataframes

Here are the steps involved in joining data using Pandas Dataframe APIs.

- First, we need to define index for both the dataframes that are supposed to be joined.
- Use join to join the Dataframes based on index. The default join is left, we need to make sure right join is used based on the requirement.

In our case, we will join orders and customers using inner join.

```python
import json
```

```python
#Create a function to get column names from json file
def get_column_names(schemas, ds_name, sorting_key='column_position'):
    column_details = schemas[ds_name]
    columns = sorted(column_details, key=lambda col: col[sorting_key])
    return [col['column_name'] for col in columns]
```

```python
schemas = json.load(open('Data/Retail_DB/schemas.json'))
```

```python
orders_column = get_column_names(schemas, 'orders')
orders_column
```

```python
customers_column = get_column_names(schemas, 'customers')
customers_column
```

```python
order = pd.read_csv(
    'Data/Retail_DB/orders.txt',
    names = orders_column  #adding column names/headers
    )
order
```

```python
customer = pd.read_csv(
    'Data/Retail_DB/customers.txt',
    names = customers_column  #adding column names/headers
    )
customer
```

```python
customer.join?
```

The join() function will join two data frames together on a key column or index. That's why, we need to use set_index() to make sure the appropriate column is changed to index and then it should work without any
issues.

```python
order = order.set_index('order_customer_id')
order
```

```python
customer = customer.set_index('customer_id')
customer
```

```python
#Join Order and Customer Dataframes
customer_order = customer.join(order, how='inner')
customer_order
```

```python
customer_order.shape
```

## Perform aggregation of Join result

```python
customer_order.head(2)
```

```python
customer_order.\
    reset_index().head(2)
```

```python
# Count of order_id by customer_id
customer_order.\
    reset_index().\
        groupby('customer_id')['order_id'].\
            agg(count_of_orderID = 'count').\
                reset_index().\
                    query('count_of_orderID == 10')
```
