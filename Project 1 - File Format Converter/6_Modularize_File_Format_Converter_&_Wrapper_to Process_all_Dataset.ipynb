```python
import pandas as pd
import glob as glb
import json
import re
import os
```

```python
def get_column_names(schemas, ds_name, sorting_keys='column_position'):
    column_details = schemas[ds_name]
    columns = sorted(column_details, key= lambda col: col[sorting_keys])
    return[col['column_name'] for col in columns]
```

```python
def read_csv(file, schemas):
    filename = re.split('[/\\\.]', file)
    ds_name = filename[-2]
    #print(filename)
    columns = get_column_names(schemas, ds_name)
    df = pd.read_csv(file, names = columns)
    return df   #, ds_name
```

```python
def to_json(df, target_base_dir, ds_name):
    json_path_name = f'{target_base_dir}/{ds_name}.json'
    os.makedirs(f'{target_base_dir}', exist_ok=True)
    df.to_json(json_path_name, orient = 'records', lines = True)
    
```

```python
def file_format_converter(src_base_dir, target_base_dir, ds_name):    
    schemas = json.load(open(f'{src_base_dir}/schemas.json'))
    src_files = glb.glob(f'{src_base_dir}/{ds_name}.txt')

    for file in src_files:
        df = read_csv(file, schemas)
        to_json(df, target_base_dir, ds_name)
```

```python
def process_files(ds_names=None):
    src_base_dir = 'Data/Retail_DB'
    target_base_dir = 'Data/Retail_DB_Jsn'
    schemas = json.load(open(f'{src_base_dir}/schemas.json'))
    
    if not ds_names:
        ds_names = schemas.keys()
    for ds_name in ds_names:
        print(f'Processing {ds_name}')
        file_format_converter(src_base_dir, target_base_dir, ds_name)
```

```python
process_files()
```
