## put together all in the below cells with Dynamic Query
- read csv file into dataframe by adding column names from json file
- create a new directory for json files and write dataframe to json files

```python
def get_column_names(schemas, ds_name, sorting_key = 'column_position'):
    column_details = schemas[ds_name]
    columns = sorted(column_details, key= lambda col: col[sorting_key])
    return[col['column_name'] for col in columns]
```

```python
schemas = json.load(open('Data\Retail_DB\schemas.json'))
```

```python
target_base_dir = 'Data\Retail_DB_json'
src_file_name = glb.glob('Data/Retail_DB/*.txt')

for file in src_file_name:
    filename = re.split('[/\\\.]', file)
    ds_name = filename[-2]
    columns = get_column_names(schemas, ds_name)
    df = pd.read_csv(file, names = columns)
    #print(df.head())
    os.makedirs(f'{target_base_dir}', exist_ok= True)
    df.to_json(f'{target_base_dir}\{ds_name}.json', orient='records', lines=True)
```
